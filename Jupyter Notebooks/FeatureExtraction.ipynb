```python
# Calculating PositiveScore , NegativeScore , PolarityScore , SubjectivityScore

def variables(text):
  PositiveScore=0
  NegativeScore=0
  tokens = word_tokenize(text)
  word_count = 0
  for word in tokens:
    if word in positive_dict['Positive_dict']:
      PositiveScore += 1
      word_count += 1
    elif word in negative_dict['Negative_dict']:
      NegativeScore += 1
      word_count += 1
    else:
      pass
  PolarityScore = (PositiveScore - NegativeScore)/((PositiveScore + NegativeScore) + 0.000001)
  totalwords= word_count
  SubjectivityScore = round((PositiveScore + NegativeScore)/((totalwords)+0.000001))
  return pd.Series([PositiveScore,NegativeScore,PolarityScore,SubjectivityScore])

```

```python
# Function to calculate total no. of clean words(by Removing Stopwords(from nltk library) and Puncuations)

def word_clean(text):
  # nltk stopwords
  stop_words = stopwords.words('english')
  # Converting text to lower case
  text1 = str(text).lower()
  # Removing Punctuations
  text1 = re.sub('\W+',' ', text1)
  words = word_tokenize(text1)
  word_count = 0
  for word in words:
    if word not in stop_words:
      word_count += 1
  return word_count

```

```python
# Function to calculate no. of complex words in each text

def complex_words_count(string):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    # Splitting the string into words

    for word in string.split():
        # Counting Syllables
        if (word.endswith('es') or word.endswith('ed')) == False:
            vowel_count = 0
            # Counting the number of vowels in each word
            for letter in word.lower():
                if letter in vowels:
                    vowel_count += 1
            # Counting complex words
            if vowel_count > 2:
                count += 1

    return count
```

```python
# Function to calculate syllable count per word (this function will return a list of values)

def syllable_count_per_word(string):
    vowels = ['a', 'e', 'i', 'o', 'u']
    syllable_count = []
    # Splitting the string into words

    for word in string.split():
        # Counting Syllables
        if (word.endswith('es') or word.endswith('ed')) == False:
            vowel_count = 0
            # Counting the number of vowels in each word
            for letter in word.lower():
                if letter in vowels:
                    vowel_count += 1
            syllable_count.append(vowel_count)
        else:
            # If word ends with 'es' or 'ed' there is no syllable i.e. 0 (non-syllable word)
            syllable_count.append(0)
    return syllable_count

```

```python
# Function to calculate Personal Pronouns in each Text

def personal_pronoun (string):
  # Country name US exlcuded and keeping only those personal pronouns in the pattern that is mentioned in the provided document(Section 7)
  personal_pronoun_pattern = r'\b(?!US\b)(?i)(I|we|my|ours|us)\b'
  personal_pronouns = re.findall(personal_pronoun_pattern, string)
  length = len(personal_pronouns)
  return length
```

```python
# Function to calculate the Average Word Length

def avg_word_length(string):
  words = string.split()
  # total_length is sum of total no. of characters in each word of the text
  total_length = 0
  total_words = 0
  for word in words:
    total_length += len(word)
    total_words += 1
  average_word_length = total_length / total_words
  return average_word_length
```

```python
# Extracting remaining features (Average_Sentence_Length,Percentage_of_Complex_Words,Fog_Index,Average_Number_of_words_per_sentence,Complex_word_count,Word_Count
#                    ,Syllable_Count_per_word,Personal_Pronouns,Average_Word_length)

# NOTE :-  1. In the below function we are dealing with two different texts :-
#          2. text1 is the data['ArticleText'] after removing all punctuations from the text
#          3. text is data['ArticleText']
#          4. This function will return a series of values

def calculate(text):
  text1 = str(text)
  text1 = re.sub('\W+',' ', text1)

  number_of_words = word_clean(text)    # Number of words
  number_of_sentences = len(str(text).split('.')) # Number of sentences
  no_of_complex_words = complex_words_count(text1) # Number of complex words
  number_of_syllable_per_word = syllable_count_per_word(text1) # Number of syllables per word (list of values)
  number_of_personal_pronouns = personal_pronoun(text1) # Number of personal pronouns
  average_word_length = avg_word_length(text1)  # Average word length

  # Calculating the features from above values
  Average_Sentence_Length = number_of_words / number_of_sentences
  Percentage_of_Complex_Words = (no_of_complex_words / number_of_words)*100
  Fog_Index = 0.4 * (Average_Sentence_Length + Percentage_of_Complex_Words)
  Average_Number_of_words_per_sentence = number_of_words / number_of_sentences
  Complex_word_count = no_of_complex_words
  Word_Count = number_of_words
  Syllable_Count_per_word = number_of_syllable_per_word
  Personal_Pronouns = number_of_personal_pronouns
  Average_Word_length = average_word_length

  return(pd.Series([Average_Sentence_Length,Percentage_of_Complex_Words,Fog_Index,Average_Number_of_words_per_sentence,Complex_word_count,Word_Count
                    ,Syllable_Count_per_word,Personal_Pronouns,Average_Word_length]))
```

```python
columns = ['AVERAGE SENTENCE LENGTH','PERCENTAGE OF COMPLEX WORDS','FOG INDEX','AVERAGE NUMBER OF WORDS PER SENTENCE','COMPLEX WORD COUNT','WORD COUNT',
           'SYLLABLE COUNT PER WORD','PERSONAL PRONOUNS','AVERAGE WORD LENGTH']

data1[columns] = data1['ArticleText'].apply(calculate)
```
